```python
import numpy as np
```

```python
path = '/content/makeSenseOfCensus.csv'
data = np.genfromtxt(path, delimiter=",", skip_header=1)

# printing the data

print("\nData: \n\n", data)
```

```python
print("\nType of data: \n\n", type(data))
```

```python
new_record = [[50, 9, 4, 1, 0, 0, 40, 0]]
np.concatenate((data, new_record), axis = 0)
```

```python
age = np.array(data[:,0])
print(age)
```

```python
max_age = np.max(age)
min_age = np.min(age)
age_mean = np.mean(age)
age_std = np.std(age)
print(max_age)
print(min_age)
print(age_mean)
print(age_std)
```

```python
race = np.array(data[:,2])
print(race)
```

```python
race_0 = np.array(race[race == 0])
race_1 = np.array(race[race == 1])
race_2 = np.array(race[race == 2])
race_3 = np.array(race[race == 3])
race_4 = np.array(race[race == 4])
print(race_0)
print(race_1)
print(race_2)
print(race_3)
print(race_4)
```

```python
len_0 =len(race_0)
len_1 =len(race_1)
len_2 =len(race_2)
len_3 =len(race_3)
len_4 =len(race_4)

print(len(race_0))
print(len(race_1))
print(len(race_2))
print(len(race_3))
print(len(race_4))
```

```python
minority_race = min(len_0, len_1, len_2, len_3, len_4)
print(minority_race)
```

```python
senior_citizens = np.array(age[age>60])
print(senior_citizens)
```

```python
working_hour_sum =np.sum(data[:,6])
print(working_hour_sum)
```

```python
senior_citizens_len = len(senior_citizens)
print(senior_citizens_len)
```

```python
avg_working_hours = working_hour_sum / senior_citizens_len
print(avg_working_hours)
```

```python
education_num = np.array(data[:,1])
print(education_num)
```

```python
high = np.array(education_num[education_num >10])
low = np.array(education_num[education_num <=10])
print(high)
print(low)
```

```python
income = np.array(data[:,7])
print(income)
```

```python
avg_pay_high = np.mean(income[education_num > 10])
print('Average pay high:',avg_pay_high)
avg_pay_low = np.mean(income[education_num <= 10])
print('Average pay low',avg_pay_low)
```
